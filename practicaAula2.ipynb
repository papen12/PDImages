```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

imagen = cv2.imread('img.jpg')

img_yuv = cv2.cvtColor(imagen, cv2.COLOR_BGR2YUV)
img_rgb = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)

fig, axs = plt.subplots(1, 3, figsize=(20, 10))

axs[0].imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
axs[0].set_title('Imagen Original')

axs[1].imshow(img_yuv)
axs[1].set_title('BGR a YUV')

axs[2].imshow(img_rgb)
axs[2].set_title('YUV a RGB')

for ax in axs:
    ax.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

gray_image = cv2.imread('profile.jpg', cv2.IMREAD_GRAYSCALE)

color_image = cv2.cvtColor(gray_image, cv2.COLOR_BAYER_BG2RGB) 

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Imagen Bayer (Grayscale)')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(color_image)
plt.title('Imagen Reconstruida (RGB)')
plt.axis('off')

plt.show()
```
