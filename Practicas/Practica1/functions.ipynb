```python
def factorial():
    n=int(input("Ingrese un numero: "))
    if n == 0: 
        print(1)
    else:
        cont = 1
        for x in range(1, n + 1):
            cont *= x
        print(cont)
    
def primo():
    n=int(input("Ingrese un numero: "))
    cont=0
    for i in range(1, n + 1):
           if(n % i==0):cont+=1
    if(cont>2): print("El numero no es primo")
    else: print("El numero es primo")
     
def mayor():
    num=[]
    for i in range (1,4):
        n=int(input(f"Ingrese un numero {i}/3: "))
        num.append(n)
    encontrarMayor(num)
    
    
def encontrarMayor(l):
    m = l[0]  
    for n in l:
        if n > m:
            m = n
    print(f"El numero mayor ingresado es {m}")
    
def palindromo():
    cad=input("Ingrese una cadena: ")
    cade=""
    for i in range(len(cad) - 1, -1, -1):
        cade+=cad[i]
    
    if(cade==cad):print(f"{cade} es un palindromo.")  
    else:print(f"{cade} no es un palindromo.")  
    
    
def repeticiones():
    num = []
    for i in range(1, 8):
        n = int(input(f"Ingrese un numero {i}/7: "))
        num.append(n)
    max = 0
    nRep = 0
    
    for n in num:
        cont = num.count(n)  
        if cont > max:
            max = cont
            nRep = n
    
    print(f"El número que más se repite es {nRep}, con {max} repeticiones.")

```
