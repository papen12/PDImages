```python
def factorial():
    n=int(input("Ingrese un numero: "))
    if n == 0: 
        print(1)
    else:
        cont = 1
        for x in range(1, n + 1):
            cont *= x
        print(cont)
    
def primo():
    n=int(input("Ingrese un numero: "))
    cont=0
    for i in range(1, n + 1):
           if(n % i==0):cont+=1
    if(cont>2): print("El numero no es primo")
    else: print("El numero es primo")
     
def mayor():
    num=[]
    for i in range (1,4):
        n=int(input(f"Ingrese un numero {i}/3: "))
        num.append(n)
    encontrarMayor(num)
    
    
def encontrarMayor(l):
    m = l[0]  
    for n in l:
        if n > m:
            m = n
    print(f"El numero mayor ingresado es {m}")
    
def palindromo():
    cad=input("Ingrese una cadena: ")
    cade=""
    for i in range(len(cad) - 1, -1, -1):
        cade+=cad[i]
    
    if(cade==cad):print(f"{cade} es un palindromo.")  
    else:print(f"{cade} no es un palindromo.")  
    
    
def repeticiones():
    num = []
    for i in range(1, 8):
        n = int(input(f"Ingrese un numero {i}/7: "))
        num.append(n)
    max = 0
    nRep = 0
    
    for n in num:
        cont = num.count(n)  
        if cont > max:
            max = cont
            nRep = n
    
    print(f"El número que más se repite es {nRep}, con {max} repeticiones.")

while True:
        print("\nSeleccione una opción:")
        print("1. Calcular Factorial")
        print("2. Verificar si es Primo")
        print("3. Encontrar el Mayor")
        print("4. Verificar si es Palíndromo")
        print("5. Contar Repeticiones")
        print("6. Salir")
        try:
            opcion = int(input("Ingrese el número de la opción: "))
            
            if opcion == 1:
                factorial()
            elif opcion == 2:
                primo()
            elif opcion == 3:
                mayor()
            elif opcion == 4:
                palindromo()
            elif opcion == 5:
                repeticiones()
            elif opcion == 6:
                print("fin del programa")
                break
            else:
                print("Opción no válida, por favor intente nuevamente.")
        except ValueError:
            print("Por favor, ingrese un número válido.")

```
