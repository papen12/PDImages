```python
import tkinter as tk
from tkinter import messagebox
from faker import Faker

faker = Faker()

def generar_nombres():
    try:
        cantidad = int(entry_cantidad.get())
    except ValueError:
        messagebox.showerror("Error", "Por favor, introduce un número válido.")
        return

    listbox_nombres.delete(0, tk.END)
    listbox_apellidos.delete(0, tk.END)
    listbox_completos.delete(0, tk.END)

    nombres = []
    apellidos = []
    completos = []

    for _ in range(cantidad):
        nombre = faker.first_name()
        apellido = faker.last_name()
        nombres.append(nombre)
        apellidos.append(apellido)
        completos.append(f"{nombre} {apellido}")

    for nombre in nombres:
        listbox_nombres.insert(tk.END, nombre)
    for apellido in apellidos:
        listbox_apellidos.insert(tk.END, apellido)
    for completo in completos:
        listbox_completos.insert(tk.END, completo)

ventana = tk.Tk()
ventana.title("Generador de Nombres")
ventana.geometry("600x400")

tk.Label(ventana, text="Cantidad de nombres a generar:").pack(pady=5)
entry_cantidad = tk.Entry(ventana)
entry_cantidad.pack(pady=5)

btn_generar = tk.Button(ventana, text="Generar Nombres", command=generar_nombres)
btn_generar.pack(pady=10)

frame_listbox = tk.Frame(ventana)
frame_listbox.pack(pady=10, fill=tk.BOTH, expand=True)

listbox_nombres = tk.Listbox(frame_listbox)
listbox_nombres.pack(side=tk.LEFT, fill=tk.BOTH, expand=True, padx=5)

listbox_apellidos = tk.Listbox(frame_listbox)
listbox_apellidos.pack(side=tk.LEFT, fill=tk.BOTH, expand=True, padx=5)

listbox_completos = tk.Listbox(frame_listbox)
listbox_completos.pack(side=tk.LEFT, fill=tk.BOTH, expand=True, padx=5)

ventana.mainloop()
```
