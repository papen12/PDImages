UMBRALIZACION POR DEFECTO TUZERO OPENC

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('img.jpg',0) #leo en escala de grises
_,umbr_trunc= cv2.threshold(imagen, 127,200, cv2.THRESH_TRUNC)
_,umbr_tozero= cv2.threshold(imagen, 127,200,cv2.THRESH_TOZERO)
_,umbr_tozero_INV=cv2.threshold(imagen,127,255,cv2.THRESH_TOZERO_INV)

plt.figure(figsize=(12,5))
plt.subplot(1,4,1)
plt.imshow(imagen, cmap='gray')
plt.title('original')

plt.subplot(1,4,2)
plt.imshow(umbr_trunc, cmap='gray')
plt.title('umbr_trunc')

plt.subplot(1,4,3)
plt.imshow(umbr_tozero, cmap='gray')
plt.title('umbr_tozero')

plt.subplot(1,4,4)
plt.title("tozero invertido")
plt.imshow(umbr_tozero_INV,cmap="gray")
plt.show()
```

CONTRASTE



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def contraste(imagen,factor=1.0):
    img=imagen.astype(np.float32)
    media=np.mean(img)
    imagen_contraste=media+(img-media)*factor
    imagen_contraste=np.clip(imagen_contraste,0,255).astype(np.uint8)
    return imagen_contraste

imagen = cv2.imread('img.jpg',0)
factor_contrast=4
img_contraste=contraste(imagen,factor_contrast)
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("original")
plt.imshow(imagen,cmap="gray")
plt.subplot(1,2,2)
plt.title("imagen contraste")
plt.imshow(img_contraste,cmap="gray")

plt.show()
```

CONTRASTE A COLORES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def contraste_color (imagen,factor=1.0):
    img=imagen.astype(np.float32)
    for i in range(3):
        media=np.mean(img[:,:,i])
        img[:,:,i]=media+(img[:,:,i]-media)*factor
        img[:,:,i]=np.clip(img[:,:,i],0,255)
    return img.astype(np.uint8)



imagen = cv2.imread('img.jpg')
factor_contrast=8
img_contraste=contraste_color(imagen,factor_contrast)
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("original")
plt.imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title("imagen contraste")
plt.imshow(cv2.cvtColor(img_contraste,cv2.COLOR_BGR2RGB))

plt.show()

plt.show()
```

CONTRASTE Y BRILLO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def ajuste(imagen, contraste=1, brillo=0):
    img=imagen.astype(np.float32)
    media=np.mean(imagen)
    img_ajustada=media+(img-media)*contraste+brillo
    img_ajustada=np.clip(img_ajustada,0,255).astype(np.uint8)
    return img_ajustada

img = cv2.imread('img.jpg')
factor_contrast=123
factor_brillo=-123
img_ajustada=ajuste(img,factor_contrast,factor_brillo)
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("original")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title("imagen brillo")
plt.imshow(cv2.cvtColor(img_ajustada,cv2.COLOR_BGR2RGB))

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def contraste(imagen,factor=1.0):
    img=imagen.astype(np.float32)
    media=np.mean(img)
    contrastada=media+(img-media)*factor
    contrastada=np.clip(contrastada,0,255).astype(np.uint8)
    return contrastada
imagen = cv2.imread('img.jpg',0)
factor_contrast=123
img_contraste=contraste(imagen,factor_contrast)
histograma_ori=cv2.calcHist([imagen],[0],None,[256],[0,256])
histograma_nueva=cv2.calcHist([img_contraste],[0],None,[256],[0,256])
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.title("hitograma original")
plt.imshow(imagen,cmap="gray")
plt.subplot(2,2,2)
plt.title("imagen hitograma nueva")
plt.plot(histograma_nueva,color="gray")
plt.xlabel("intencidad")
plt.ylabel("frecuencia")




plt.subplot(2,2,3)
plt.title("imagen histograma nueva")
plt.imshow(img_contraste,cmap="gray")

plt.subplot(2,2,4)
plt.title("imagen histograma nueva")
plt.plot(histograma_nueva,color="gray")
plt.xlabel("intensidad")
plt.ylabel("frecuencia")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams ['figure.figsize']=[10,5]
imagen = cv2.imread('img.jpg')
imagen_rgb= cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

#MOSTRANDO UNA IMAGE
plt.imshow(imagen_rgb)
plt.show()

#seleccionando un punto de interes

plt.imshow(imagen_rgb[50:100,20:65,:])

plt.show()

#guardado de imagen seleccionada
img_cortada=imagen_rgb[50:100,20:65,:]
plt.imsave('img_cortada.jpg', img_cortada)
#histograma de canales de la imagen seccionada
seccionada=cv2.imread('img_cortada.jpg')
color=('r','g','b')
for canal, col in enumerate(color):
    histograma=cv2.calcHist ([seccionada*255], [canal], None, [256], [0,256])
    plt.plot(histograma, color=col)
    plt.xlim([0,256])
plt.show()
```

MASCARA

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer la imagen
imagen = cv2.imread('img.jpg')

if imagen is None:
    print("Error: No se pudo cargar la imagen.")
else:
    # Convertir BGR a RGB para matplotlib
    imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

    # Dimensiones de la imagen
    alto, ancho = imagen_rgb.shape[:2]

    # Crear máscara circular
    mascara = np.zeros((alto, ancho), dtype=np.uint8)  # Máscara de un solo canal
    centro = (ancho // 2, alto // 2)
    radio = min(ancho, alto) // 4  # Radio como un cuarto del lado más corto
    cv2.circle(mascara, centro, radio, (255), -1)

    # Aplicar la máscara
    img_mascara = cv2.bitwise_and(imagen_rgb, imagen_rgb, mask=mascara)

    # Mostrar las imágenes
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(imagen_rgb)
    plt.title('Imagen Original')

    plt.subplot(1, 2, 2)
    plt.imshow(img_mascara)
    plt.title('Imagen con Máscara Circular')

    plt.show()
```
